```python
# Código executado e mostrado para o usuário: plota uma dispersão "toy model" para uma equação de Dirac fracionária em 1D.
# Observações:
# - Isto é um modelo fenomenológico didático: substituímos as derivadas por potências fracionárias no espaço (|k|^α)
#   e construímos um Hamiltoniano 2x2 estilo Dirac (com Pauli) cuja autofunção dá níveis E = ± sqrt((|k|^α)^2 + (m^α/α)^2).
import numpy as np
import matplotlib.pyplot as plt

# parâmetros
m = 1.0           # massa em unidades arbitrárias
ks = np.linspace(-10, 10, 800)  # variação do momento
alphas = [1.0, 0.9, 0.7, 0.5]   # valores de alfa para comparar

plt.figure(figsize=(9,6))
for alpha in alphas:
    # modelo toy: autovalores E = ± sqrt( (|k|^α)^2 + (m^α/α)^2 )
    # uso |k|^alpha para preservar paridade; a divisão por alpha é o fator que apareceu na sua equação
    Ek = np.sqrt((np.abs(ks)**alpha)**2 + (m**alpha / alpha)**2)
    plt.plot(ks, Ek, label=f"α = {alpha:.2f}")
    plt.plot(ks, -Ek, linestyle='--')

plt.axhline(0, linewidth=0.7)
plt.title("Dispersão do modelo brinquedo: 'Dirac fracionário' em 1D (ilustrativo)")
plt.xlabel("k (momento)")
plt.ylabel("E(k)")
plt.legend(title="Parâmetros")
plt.grid(True)
plt.tight_layout()

# Salva imagem para download
#plt.savefig("/mnt/data/dispersion_dirac_fracional_toy.png", dpi=200)
plt.show()

#print("Figura gerada e salva em /mnt/data/dispersion_dirac_fracional_toy.png")
```
