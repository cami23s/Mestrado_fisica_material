```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# Parâmetros físicos
hbar = 1.0
n = 1.0
omega = 1.0
E_n = 0.5 * hbar * omega  # Energia do estado fundamental

# Tempo
t_vals = np.linspace(0.1, 11, 300)

# Valores de alpha
alphas = [0.6, 0.8, 1]
colors = ['orange', 'green', 'navy']
styles = ['--', '-', ':']
labels = [r'$\alpha = 0.6$', r'$\alpha = 0.8$', r'$\alpha = 1$']

# Função de Mittag-Leffler - apenas parte real com cosseno
def mittag_leffler_real(alpha, z, terms=300):
    result = 0
    for n in range(terms):
        cos_term = np.cos(np.pi * n * alpha / 2)
        result += cos_term * (z**n) / gamma(alpha * n + 1)
    return result

# Gráfico
plt.figure(figsize=(8, 5))

for alpha, color, style, label in zip(alphas, colors, styles, labels):
    g_t = []
    for t in t_vals:
        z = (E_n / hbar**alpha) * t**alpha
        g_val = mittag_leffler_real(alpha, z)
        g_t.append(g_val)

    plt.plot(t_vals, g_t, style, color=color, linewidth=2, label=label)

# Estilo do gráfico
plt.xlabel(r'$t$', fontsize=12)
plt.xlim(0,10)
plt.ylabel(r'$E_\alpha(t)$', fontsize=14)
plt.ylim(-1.5, 2)
plt.title(r'g(t) - Oscilador Harmônico', fontsize=14)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Pacotes de Onda e Função de Green**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# --------------------------
# Parâmetros físicos
# --------------------------
hbar = 1.0
m = 1.0

# Valor de k fixo (não zero)
k = 2.0

# Intervalo de tempo
t_vals = np.linspace(0, 5, 300)

# Valores de alpha
alphas = [0.6, 0.8, 1.0]
colors = ['orange', 'green', 'navy']
styles = ['--', '-', ':']
labels = [r'$\alpha = 0.6$', r'$\alpha = 0.8$', r'$\alpha = 1$']

# --------------------------
# Função Mittag-Leffler (parte real)
# --------------------------
def mittag_leffler_real(alpha, z, terms=200):
    result = 0.0 + 0j
    for n in range(terms):
        result += (z**n) / gamma(alpha * n + 1)
    return np.real(result)

# --------------------------
# Cálculo da função E_alpha(z)
# --------------------------
plt.figure(figsize=(8,5))

for alpha, color, style, label in zip(alphas, colors, styles, labels):
    g_t = []
    for t in t_vals:
        z = ((-1j * hbar * t) / (2 * m))**alpha * (k**2)
        g_val = mittag_leffler_real(alpha, z)
        g_t.append(g_val)

    g_t = np.array(g_t)
    plt.plot(t_vals, g_t, style, color=color, linewidth=2, label=label)

# --------------------------
# Gráfico
# --------------------------
plt.title(r'$g(t)$ — Função de Green Fracionária', fontsize=14)
plt.xlabel(r'$t$', fontsize=12)
plt.ylabel(r'$E_\alpha(t)$', fontsize=14)
plt.xlim(0, 3)
plt.ylim(-2, 3)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
